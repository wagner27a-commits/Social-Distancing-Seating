## Single with Built In Methods

```python
L = [1,3]
for x in range(1,1000):
    if x%2==1:
        for y in range(0,((x-3)/2).floor()+1):
            for z in range(0,x-2):
                if x==2*y+z+3 and is_power_of_two(y+1) and is_power_of_two(z):
                    if x in L:
                        break
                    else:
                        L.append(x)
                        break
for i in L:
    print(i)
```

# Watering Hole Simulations

## Double Input

```python
n = Integer(input("n:"))
W = [0 for i in range (n)]
f = Integer(input("first position: "))
W[f-1] = 1

for k in range(2,(n/2).ceil()+1):
    L = [0 for i in range (n)]
    S = [0 for i in range (n)]
    for i in range(0,n):
        M = [0 for i in range (n)]
        for j in range(0,n):
            if j==i and W[j]!=0:
                M[j]=-1
            elif j!=i and W[j]!=0:
                M[j]=abs(j-i)
                
        Q = []
        for z in range(0,n):
            if M[z]>=2:
                Q.append(M[z])
        if -1 in M or 1 in M or len(Q)==0:
            L[i]=-1+(-1*n)
        elif len(Q)==1:
            L[i]=max(M)
        else:
            l = Q.index(min(Q))
            L[i] = min(Q)
            Q.pop(Q.index(min(Q)))
            S[i] = min(Q)
            
    x = L.index(max(L))
    if k!=2:
        for u in range(0,n):
            if x!=u and L[u]==max(L) and S[u] < S[x]:
                x = u
        if max(L)>=0:
            f2 = x
            W[f2]=k
        else:
            break
    else:
        if max(L)>=0:  
            f2 = x
            W[f2]=k
        else:
            break
print(W)
if max(W)==(n/2).ceil():
    print(n, f, " works!")
```
